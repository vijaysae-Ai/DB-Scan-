```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets 
```

```python
x,y= datasets.make_circles(n_samples=1500,noise=0.05,factor=0.05,
   random_state=10)
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN( eps=0.1,min_samples=3)
```

```python
db.fit(x)
```

```python
y_pred= db.fit_predict(x)
```

```python
y_pred
```

```python
label=db.labels_
```

```python
label
```

```python
plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.scatter(marker="*",c='r')
```
